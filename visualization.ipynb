```python
def show_countplot(x, hue, ax, data):
    sns.countplot(x=x, hue=hue, data=data, ax=ax)

def show_target_label_plot(label, show_row):
    show_countplot(label, 'treatment', ax[show_row][0], fdata)
    

fig, ax = plt.subplots(nrows=13, ncols=1, figsize=(24,90))

show_target_label_plot('Age', 0)
show_target_label_plot('Gender', 1)
show_target_label_plot('Country', 2)

show_target_label_plot('self_employed', 3)
show_target_label_plot('family_history', 4)
show_target_label_plot('work_interfere', 6)


show_target_label_plot('remote_work', 7)
show_target_label_plot('tech_company', 8)
show_target_label_plot('benefits', 9)

show_target_label_plot('mental_health_consequence', 10)
show_target_label_plot('phys_health_consequence', 11)
show_target_label_plot('mental_vs_physical', 12)
```
